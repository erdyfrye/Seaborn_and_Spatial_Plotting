### 2. Distribusi Plot

```python
### 1. Distplot

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['tip'])
```

```python
tips.describe()
```

```python
### 2. Joint Plot

sns.jointplot(data=tips, x=tips['total_bill'], y=tips['tip'])
```

```python
### 3. Pair Plot
sns.pairplot(data=tips)
```

### 3. Categorical Plot

```python
#barplot

sns.barplot(data=tips, x=tips['sex'], y=tips['tip'])
```

```python
#boxplot

sns.boxplot(data=tips, x=tips['sex'], y=tips['total_bill'], hue=tips['time'])
```

```python
sns.boxplot(data=tips, x=tips['sex'], y=tips['tip'], hue=tips['smoker'])
```

```python
#violin plot

sns.violinplot(data=tips, x=tips['sex'], y=tips['total_bill'], hue=tips['time'])
```

```python
#strip plot

sns.stripplot(data=tips, x=tips['sex'], y=tips['total_bill'], hue=tips['time'])
```

```python
import folium
```

```python
telu
```

```python
from branca.element import Figure
fig = Figure(width = 600, height = 300)
```

```python
telu = folium.Map(location = [-6.971356322719667, 107.63128843092986], min_zoom=0, max_zoom=18, zoom_start=15)
fig.add_child(telu)
telu
```

```python
fig2=Figure(width=550,height=350)
m2=folium.Map(location=[-6.9730598, 107.6295034],zoom_start=14,min_zoom=8,max_zoom=16)
fig2.add_child(m2)
folium.TileLayer('Stamen Terrain').add_to(m2)
folium.TileLayer('Stamen Toner').add_to(m2)
folium.TileLayer('Stamen Water Color').add_to(m2)
folium.TileLayer('cartodbpositron').add_to(m2)
folium.TileLayer('cartodbdark_matter').add_to(m2)
folium.LayerControl().add_to(m2)
m2
```

```python
folium.Marker(location=[-6.971356322719667, 107.63128843092986], popup='Telkom_University').add_to(m2)
folium.Marker(location=[-6.970003832100403, 107.62753333845129], popup='Asrama_TelU', icon=folium.Icon(color='read', icon='none')).add_to(m2)
m2
```

```python
import pandas as pd
```

```python
bikesharing = pd.read_csv('dublin.csv')
```

```python
bikesharing.head()
```

```python
bikesharing.describe()
```

```python
bikesharing.info()
```

```python
bikesharing.drop_duplicates()
```

```python
lat=bikesharing['Latitude'].mean()
long=bikesharing['Longitude'].mean()
```

```python
dublin=folium.Map(location=[lat, long], min_zoom=0, max_zoom=18, zoom_start=15)
dublin
```

```python
for i, row in bikesharing.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=row['Address']).add_to(dublin)
```

```python
dublin
```

```python
dublin.save('dublin.html')
```
